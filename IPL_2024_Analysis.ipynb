 ## Toss Decision Analysis


```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='decision')
plt.title('Toss Decision Analysis')
plt.xlabel('Decision (Bat/Field)')
plt.ylabel('Count')
plt.show()
```

 *Shows how often teams chose to bat or field after winning the toss.*

 ## Match Outcome by Toss Decision

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='decision', hue='winner')
plt.title('Match Outcome by Toss Decision')
plt.xlabel('Decision (Bat/Field)')
plt.ylabel('Match Count')
plt.show()
```

*Displays the relationship between the toss decision (bat/field) and match outcomes.*

## Scores Comparison

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='first_score', y='second_score', hue='winner', style='decision', s=100)
plt.title('First Innings Scores vs Second Innings Scores')
plt.xlabel('First Innings Score')
plt.ylabel('Second Innings Score')
plt.show()
```

 *Compares first innings scores with second innings scores.*

## Winning Margins

```python
df['Winning_Margin'] = abs(df['first_score'] - df['second_score'])
plt.figure(figsize=(20, 6))
sns.barplot(data=df, x='id', y='Winning_Margin', hue='winner')
plt.title('Winning Margins by Match')
plt.xlabel('Match ID')
plt.ylabel('Winning Margin (Runs/Wickets)')
plt.show()
```

*Highlights the margin of victory in each match.*

## Top Scorers

```python
plt.figure(figsize=(20, 12))
sns.barplot(data=df, x='id', y='first_score', hue='most_runs')
plt.title('Top Scorers in Each Match')
plt.xlabel('Match ID')
plt.ylabel('Runs Scored')
plt.show()
```

*Identifies the top run-scorers in each match.*

## Top Wicket-Takers

```python
plt.figure(figsize=(20, 12))
sns.barplot(data=df, x='id', y='first_wkts', hue='most_wkts')
plt.title('Top Wicket-Takers in Each Match')
plt.xlabel('Match ID')
plt.ylabel('Wickets Taken')
plt.show()
```

*Shows the top wicket-takers in each match.*

## Runs vs Wickets Analysis

```python
plt.figure(figsize=(20, 8))
sns.lineplot(data=df, x='first_score', y='first_wkts', marker='o', label='First Innings')
sns.lineplot(data=df, x='second_score', y='second_wkts', marker='o', label='Second Innings')
plt.title('Runs vs Wickets Analysis')
plt.xlabel('Runs')
plt.ylabel('Wickets')
plt.legend()
plt.show()
```

*Analyzes the relationship between runs scored and wickets lost.*

## Player of the Match Distribution

```python
plt.figure(figsize=(10, 15))
sns.countplot(data=df, y='player_of_the_match')
plt.title('Player of the Match Distribution')
plt.xlabel('Count')
plt.ylabel('Player of the Match')
plt.show()
```

*Distribution of "Player of the Match" awards.*

## Team-wise Performance

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='winner')
plt.title('Team-wise Performance')
plt.xlabel('Match Wins')
plt.ylabel('Team')
plt.show()
```

*Displays the number of matches won by each team.*

## Score Distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['first_score'], bins=5, kde=True, color='blue', label='First Innings')
sns.histplot(df['second_score'], bins=5, kde=True, color='orange', label='Second Innings')
plt.title('Score Distribution in First and Second Innings')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

*Histogram of first and second innings scores.*

## Wicket Distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['first_wkts'], bins=5, kde=True, color='blue', label='First Innings')
sns.histplot(df['second_wkts'], bins=5, kde=True, color='orange', label='Second Innings')
plt.title('Wicket Distribution in First and Second Innings')
plt.xlabel('Wickets')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

*Histogram of wickets lost in the first and second innings.*

##  Performance of Toss-Winning Teams

```python
toss_win_match_win = df[df['toss_winner'] == df['winner']].shape[0]
toss_win_match_loss = df[df['toss_winner'] != df['winner']].shape[0]
plt.figure(figsize=(8, 8))
plt.pie([toss_win_match_win, toss_win_match_loss], labels=['Toss Winner Won', 'Toss Winner Lost'], autopct='%1.1f%%')
plt.title('Performance of Toss-Winning Teams')
plt.show()
```

*Pie chart showing the percentage of matches won by toss-winning teams.*

 ## Head-to-Head Performance

```python
df_heatmap = pd.crosstab(df['team1'], df['team2'])
plt.figure(figsize=(10, 6))
sns.heatmap(df_heatmap, annot=True, cmap="YlGnBu")
plt.title('Head-to-Head Performance')
plt.xlabel('Team 2')
plt.ylabel('Team 1')
plt.show()
```

*Heatmap showing the head-to-head match results between teams.*
